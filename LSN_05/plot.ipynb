```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
#%matplotlib notebook
```

# Exercise 05

We are computing the estimated radius of the position of an electron in a hydrogen atom. The Schr√∂dinger-equation is evaluated with a Montecarlo-algorithm.

Displaying the s-orbit:

```python
so = np.loadtxt("s-orbit.dat")
po = np.loadtxt("p-orbit.dat")

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(so[:,0], so[:,1], so[:,2], c=so[:,2], marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)

plt.show()
```

Displaying the p-orbit:

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(po[:,0], po[:,1], po[:,2], c=po[:,2], marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

Displaying the average distances for both orbits:

```python
rs = np.loadtxt("s-rad.dat")
rp = np.loadtxt("p-rad.dat")

n = np.size(rs[:,0])
x = np.array(range(n))

plt.figure(figsize = (9,4))

plt.subplot(121)
plt.errorbar(x, rs[:,0], yerr = rs[:,1])
plt.hlines(xmin = 0, xmax = n, y = 1.5, linewidth = 1, linestyle = "dashed")
plt.xlabel("Number of Blocks")
plt.ylabel("$a_0$")
plt.title("Expectation for $r$  in s-orbit")

plt.subplot(122)
plt.errorbar(x, rp[:,0], yerr = rp[:,1])
plt.hlines(xmin = 0, xmax = n, y = 5, linewidth = 1, linestyle = "dashed")
plt.xlabel("Number of Blocks")
plt.ylabel("$a_0$")
plt.title("Expectation for $r$  in p-orbit")

plt.show()
```
