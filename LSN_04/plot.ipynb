```python
import numpy as np
from matplotlib import pyplot as plt
```

# Exercise 04

```python
eps_a = 120
sig_a = 0.34
eps_k = 164
sig_k = 0.364

k_boltzmann = 1
#k_boltzmann = 1.38 * 10**(-23)
```

The simulation returns values in reduced units. The values for Argon and Krypton can be computed later with the same set of data.

Argon:

```python
#argon solid phase:

ekin = np.loadtxt('sol_ek.dat') * eps_a * k_boltzmann
epot = np.loadtxt('sol_ep.dat') * eps_a * k_boltzmann
etot = np.loadtxt('sol_et.dat') * eps_a * k_boltzmann
temp = np.loadtxt('sol_te.dat') * eps_a
pres = np.loadtxt('sol_pr.dat') * 10**(27) * eps_a * k_boltzmann / sig_a**3

x = np.arange(ekin[:,0].size)

print('Argon in solid phase')

plt.figure( figsize = (15,15) )

plt.subplot(321)
plt.errorbar(x,etot[:,0], yerr = etot[:,1])
#plt.plot(epot[:,0] + ekin[:,0])
plt.title('Total Energy per Particle')
plt.ylabel('Energy in J')
plt.xlabel('Number of Blocks')

plt.subplot(322)
plt.errorbar(x, epot[:,0], yerr = epot[:,1])
plt.title('Potential Energy per Particle')
plt.ylabel('Energy in J')
plt.xlabel('Number of Blocks')

plt.subplot(323)
plt.errorbar(x, ekin[:,0], yerr = ekin[:,1])
plt.title('Kinetic Energy per Particle')
plt.ylabel('Energy in J')
plt.xlabel('Number of Blocks')

plt.subplot(324)
plt.errorbar(x, temp[:,0], yerr = temp[:,1])
plt.title('Temperature')
plt.ylabel('Temperature in K')
plt.xlabel('Number of Blocks')

plt.subplot(325)
plt.errorbar(x, pres[:,0], yerr = pres[:,1])
plt.title('Pressure')
plt.ylabel('Pressure in Pa')
plt.xlabel('Number of Blocks')


plt.show()
```

```python
#argon liquid phase:

ekin = np.loadtxt('liq_ek.dat') * eps_a * k_boltzmann
epot = np.loadtxt('liq_ep.dat') * eps_a * k_boltzmann
etot = np.loadtxt('liq_et.dat') * eps_a * k_boltzmann
temp = np.loadtxt('liq_te.dat') * eps_a
pres = np.loadtxt('liq_pr.dat') * 10**(27) * 1.38 * 10**(-23) * 10**(19) * eps_a * k_boltzmann / sig_a**3

print('Argon in liquid phase')

plt.figure( figsize = (15,15) )

plt.subplot(321)
plt.errorbar(x,etot[:,0], yerr = etot[:,1])
plt.title('Total Energy per Particle')
plt.ylabel('Energy in J')
plt.xlabel('Number of Blocks')

plt.subplot(322)
plt.errorbar(x, epot[:,0], yerr = epot[:,1])
plt.title('Potential Energy per Particle')
plt.ylabel('Energy in J')
plt.xlabel('Number of Blocks')

plt.subplot(323)
plt.errorbar(x, ekin[:,0], yerr = ekin[:,1])
plt.title('Kinetic Energy per Particle')
plt.ylabel('Energy in J')
plt.xlabel('Number of Blocks')

plt.subplot(324)
plt.errorbar(x, temp[:,0], yerr = temp[:,1])
plt.title('Temperature')
plt.ylabel('Temperature in K')
plt.xlabel('Number of Blocks')

plt.subplot(325)
plt.errorbar(x, pres[:,0], yerr = pres[:,1])
plt.title('Pressure')
plt.ylabel('Pressure in Pa')
plt.xlabel('Number of Blocks')


plt.show()
```
