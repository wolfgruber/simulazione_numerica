```python
import numpy as np
import numpy.random as rd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
```

```python
def f(y,x):
    return np.sin( np.power(x,2) + np.power(y,2) )
    #return 0.5 * np.power(x,2) + np.power(y,2)
```

```python
nrp1 = 30

x1 = np.linspace(-3/2,3/2,nrp1)
y1 = x1

z1 = np.empty((nrp1,nrp1))

for i in range(nrp1):
    for j in range(nrp1):
        z1[i,j] = f(x1[i], y1[j]) + rd.normal(0,0.1)
        
nrp2 = 30

x2 = np.linspace(-3/2,3/2,nrp2)
y2 = x2

z2 = np.empty((nrp2,nrp2))

for i in range(nrp2):
    for j in range(nrp2):
        z2[i,j] = f(x2[i], y2[j])
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt

xp1, yp1 = np.meshgrid(x1,y1)
xp2, yp2 = np.meshgrid(x2,y2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Grab some test data.
#X, Y, Z = axes3d.get_test_data(0.5)

# Plot a basic wireframe.
#ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

#ax.plot_surface(xp1,yp1,z1,cmap='viridis')
ax.plot_surface(xp2,yp2,z2,cmap='viridis')

#ax.plot_wireframe(xp2,yp2,z2)
#ax.plot_wireframe(xp1,yp1,z1)

#ax.scatter(xp1,yp1,z1)


plt.show()
```

```python
plt.contour(xp2,yp2,z2)
plt.show()
```

```python
# creating some training and validation data

n_train = 10000
n_valid = 100

x_train = rd.rand(n_train,2) * 3 - 1.5
x_valid = rd.rand(n_valid,2) * 3 - 1.5

y_train = np.empty(n_train)
y_valid = np.empty(n_valid)

y_train = f(x_train[:,0],x_train[:,1])
y_valid = f(x_valid[:,0],x_valid[:,1])
```

```python
# constructing a NN

fit2D = Sequential()

fit2D.add(Dense(100, input_shape = (2,)))

fit2D.add(Dense(100))
fit2D.add(Dense(100))
fit2D.add(Dense(100, activation='elu'))
fit2D.add(Dense(100, activation='elu'))
fit2D.add(Dense(100, activation='relu'))
fit2D.add(Dense(100, activation='relu'))
fit2D.add(Dense(100, activation='tanh'))
fit2D.add(Dense(100, activation='tanh'))
fit2D.add(Dense(100, activation='tanh'))
fit2D.add(Dense(100, activation='tanh'))
fit2D.add(Dense(100, activation='elu'))
fit2D.add(Dense(100, activation='elu'))
fit2D.add(Dense(100, activation='elu'))

fit2D.add(Dense(1))

fit2D.compile(optimizer='Adam', loss='mse', metrics=['mse'])
fit2D.summary()
    
hist = fit2D.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=10,
          shuffle=True,
          validation_data=(x_valid, y_valid))
```

```python
n_predict = 50

xy = np.linspace(-3/2,3/2,n_predict)
x_predict = np.transpose(np.array([xy,xy]))
y_predict = np.zeros((n_predict, n_predict))

for i in range(n_predict):
    y_predict[i] = np.reshape(fit2D.predict(np.transpose(np.array([xy,np.ones(n_predict)*i*3/n_predict-3/2]))), n_predict)

xp, yp = np.meshgrid(xy, xy)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(xp, yp, y_predict, cmap='viridis')
plt.show()
```
