```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 4 - 3*x - 2*np.power(x,2) + 3*np.power(x,3)

def rete(n,m,act):
    model = Sequential()
    model.add(Dense(m[0], input_shape=(1, ), activation=act[0]))
    
    for i in range(1,n):
        model.add(Dense(m[i] , activation =act[i]))

    model.add(Dense(1))

    model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
    model.summary()
    return model
```

```python
def evaluation(models, hist):
    n = len(models)
    m = 100
    x_predicted = np.random.uniform(-1, 1, m)
    x_predicted.sort()
    
    y_predicted = np.zeros((n,m))
    y_eval = f(x_predicted)
    
    plt.figure(figsize = (12,4*n))
    
    for i in range(n):
        y_predicted[i,:] = np.reshape(models[i].predict(x_predicted), m)
        
        score = models[i].evaluate(x_predicted, y_eval, batch_size=32, verbose=1)

        plt.subplot(n,2,1+2*i)
        plt.plot(x_predicted, y_predicted[i,:])
        plt.scatter(x_predicted, y_predicted[i,:],color='r')
        plt.plot(x_valid, y_target)
        #plt.xlim(0.5,1)
        #plt.ylim(1.8,2.5)
        plt.grid(True)

        plt.subplot(n,2,2+2*i)
        plt.plot(hist[i].history['loss'])
        plt.plot(hist[i].history['val_loss'])
        plt.title('Model loss')
        plt.ylabel('Loss')
        #plt.xlabel('Epoch')
        plt.legend(['Train', 'Test'], loc='best')

    plt.show()
```

```python
#producing training and validation data

Nvalid = 100
Ntrain = 10000

x_train = np.random.uniform(-1, 1, Ntrain)
x_valid = np.random.uniform(-1, 1, Nvalid)
x_valid.sort()

y_target = f(x_valid)

sigma = 0.00
y_train = np.random.normal(f(x_train), sigma) 
y_valid = np.random.normal(f(x_valid), sigma)

plt.plot(x_valid, y_target)
plt.plot(x_valid, y_valid,'o')

plt.grid(True); plt.show()
```

```python
#creating the network

import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
```

```python
#only linear (no activation function)

fit1 = Sequential()

fit1.add(Dense(6, input_shape = (1,), activation='relu'))

fit1.add(Dense(6))
fit1.add(Dense(6))
fit1.add(Dense(6))
fit1.add(Dense(6))
fit1.add(Dense(6))

fit1.add(Dense(1))

fit1.compile(optimizer='Adam', loss='mse', metrics=['mse'])
fit1.summary()
    
hist1 = fit1.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=10,
          shuffle=True,
          validation_data=(x_valid, y_valid))
```

```python
fit2 = Sequential()

fit2.add(Dense(6, input_shape = (1,)))

fit2.add(Dense(6))
fit2.add(Dense(6, activation='relu'))
fit2.add(Dense(6, activation='selu'))
fit2.add(Dense(6, activation='selu'))
fit2.add(Dense(6, activation='selu'))

fit2.add(Dense(1))

fit2.compile(optimizer='sgd', loss='mse', metrics=['mse'])
fit2.summary()
    
hist2 = fit2.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=10,
          shuffle=True,
          validation_data=(x_valid, y_valid))
```

```python
# something really non-linear

fit3 = Sequential()

fit3.add(Dense(6, input_shape = (1,)))

fit3.add(Dense(6))
fit3.add(Dense(6, activation='tanh'))
fit3.add(Dense(6, activation='tanh'))
fit3.add(Dense(6, activation='tanh'))
fit3.add(Dense(6, activation='tanh'))

fit3.add(Dense(1))

fit3.compile(optimizer='sgd', loss='mse', metrics=['mse'])
fit3.summary()
    
hist3 = fit3.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=10,
          shuffle=True,
          validation_data=(x_valid, y_valid))
```

```python
# a little bit bigger

fit4 = Sequential()

fit4.add(Dense(64, input_shape = (1,)))

fit4.add(Dense(64))
fit4.add(Dense(64, activation='relu'))
fit4.add(Dense(64, activation='relu'))
fit4.add(Dense(64, activation='relu'))
fit4.add(Dense(64, activation='relu'))
fit4.add(Dense(64, activation='relu'))

fit4.add(Dense(1))

fit4.compile(optimizer='sgd', loss='mse', metrics=['mse'])
fit4.summary()
    
hist4 = fit4.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=10,
          shuffle=True,
          validation_data=(x_valid, y_valid))
```

```python
# bigger and deeper
# Adam is used for optimisation, otherwise such big networks need a lot of time

fit5 = Sequential()

fit5.add(Dense(64, input_shape = (1,)))

fit5.add(Dense(64))
fit5.add(Dense(64))
fit5.add(Dense(100, activation='relu'))
fit5.add(Dense(100, activation='relu'))
fit5.add(Dense(100, activation='relu'))
fit5.add(Dense(100, activation='relu'))
fit5.add(Dense(100, activation='relu'))
fit5.add(Dense(100, activation='relu'))
fit5.add(Dense(100, activation='relu'))
fit5.add(Dense(100, activation='relu'))
fit5.add(Dense(100, activation='relu'))
fit5.add(Dense(100, activation='relu'))

fit5.add(Dense(1))

fit5.compile(optimizer='Adam', loss='mse', metrics=['mse'])
fit5.summary()
    
hist5 = fit5.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=10,
          shuffle=True,
          validation_data=(x_valid, y_valid))
```

```python
# something to work with
# not really deep but quite wide

fit6 = Sequential()

fit6.add(Dense(200, input_shape = (1,), activation='relu'))

fit6.add(Dense(200, activation='relu'))
fit6.add(Dense(200))
fit6.add(Dense(200, activation='sigmoid'))
fit6.add(Dense(200, activation='sigmoid'))
fit6.add(Dense(200, activation='sigmoid'))
fit6.add(Dense(200, activation='relu'))


fit6.add(Dense(1))

fit6.compile(optimizer='Adam', loss='mse', metrics=['mse'])
fit6.summary()
    
hist6 = fit6.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=10,
          shuffle=True,
          validation_data=(x_valid, y_valid))
```

```python
# evaluate models

evaluation([fit1, fit2, fit3, fit4, fit5, fit6], [hist1, hist2, hist3, hist4, hist5, hist6])
```

# Exercise 11.2:

```python
def f(y,x):
    return np.sin( np.power(x,2) + np.power(y,2) )


#plotting the desired function

nrp2 = 50

x2 = np.linspace(-3/2,3/2,nrp2)
y2 = x2

z2 = np.empty((nrp2,nrp2))

for i in range(nrp2):
    for j in range(nrp2):
        z2[i,j] = f(x2[i], y2[j])
        
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt

xp2, yp2 = np.meshgrid(x2,y2)

fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(121, projection='3d')

ax.plot_surface(xp2,yp2,z2,cmap='viridis')

plt.subplot(122)
plt.contour(xp2,yp2,z2)
plt.show()
```

```python
# creating some training and validation data

n_train = 10000
n_valid = 100

x_train = np.random.rand(n_train,2) * 3 - 1.5
x_valid = np.random.rand(n_valid,2) * 3 - 1.5

y_train = np.empty(n_train)
y_valid = np.empty(n_valid)

y_train = f(x_train[:,0],x_train[:,1])
y_valid = f(x_valid[:,0],x_valid[:,1])
```

```python
# constructing a NN
# 

fit2D = Sequential()

fit2D.add(Dense(100, input_shape = (2,)))

fit2D.add(Dense(100))
fit2D.add(Dense(100))
fit2D.add(Dense(100, activation='elu'))
fit2D.add(Dense(100, activation='elu'))
fit2D.add(Dense(100, activation='relu'))
fit2D.add(Dense(100, activation='relu'))
fit2D.add(Dense(100, activation='tanh'))
fit2D.add(Dense(100, activation='tanh'))
fit2D.add(Dense(100, activation='tanh'))
fit2D.add(Dense(100, activation='tanh'))
fit2D.add(Dense(100, activation='elu'))
fit2D.add(Dense(100, activation='elu'))
fit2D.add(Dense(100, activation='elu'))

fit2D.add(Dense(1))

fit2D.compile(optimizer='Adam', loss='mse', metrics=['mse'])
fit2D.summary()
    
hist = fit2D.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=10,
          shuffle=True,
          validation_data=(x_valid, y_valid))
```

```python
n_predict = 50

xy = np.linspace(-3/2,3/2,n_predict)
x_predict = np.transpose(np.array([xy,xy]))
y_predict = np.zeros((n_predict, n_predict))

for i in range(n_predict):
    y_predict[i] = np.reshape(fit2D.predict(np.transpose(np.array([xy,np.ones(n_predict)*i*3/n_predict-3/2]))), n_predict)

xp, yp = np.meshgrid(xy, xy)
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(121, projection='3d')

ax.plot_surface(xp, yp, y_predict, cmap='viridis')

bx = fig.add_subplot(122)
bx.contour(xy,xy,y_predict)
plt.show()
```
