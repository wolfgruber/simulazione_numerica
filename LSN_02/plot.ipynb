```python
import numpy as np
from matplotlib import pyplot as plt
```

# Exercise 02.1:

computing the Integral $$ I =\int_0^1 \frac{\pi}{2} \cos\left(\frac{x~\pi}{2} \right) \text{d}x = 1$$

```python
i1 = np.loadtxt("int1.dat")
i2 = np.loadtxt("int2.dat")

plt.figure(figsize = (15,5))

plt.subplot(121)
plt.plot(i1[:,0], i1[:,1], label = "Integral")
plt.plot(i1[:,0], i1[:,1]+i1[:,2], linestyle = "dashed", linewidth = 1, color = "black", label = "$\pm \sigma$")
plt.plot(i1[:,0], i1[:,1]-i1[:,2], linestyle = "dashed", linewidth = 1, color = "black")
plt.plot([0, i1[-1,0]], [1,1])
plt.title("Integrating with Uniform Distribution")
plt.legend()
plt.ylim(0.9,1.1)
#plt.hist(i2[:,3])

plt.subplot(122)
plt.plot(i2[:,0], i2[:,1], label = "Integral")
plt.plot(i2[:,0], i2[:,1]+i2[:,2], linestyle = "dashed", linewidth = 1, color = "black", label = "$\pm \sigma$")
plt.plot(i2[:,0], i2[:,1]-i2[:,2], linestyle = "dashed", linewidth = 1, color = "black")
plt.plot([0, i2[-1,0]], [1,1])
plt.title("Integrating with Importance Sampling")
plt.ylim(0.9,1.1)
plt.legend()


plt.show()
```

# Exercise 02.2

computing the mean distance coverd with a random walk

```python
dat = np.loadtxt("dat2a.dat")
dat2 = np.loadtxt("dat2b.dat")
#print(dat2)

plt.figure(figsize = (15,5))

plt.subplot(121)
#plt.plot(dat[:,0], dat[:,2]/dat[:,1], label = "relative error") #for plotting the relative error
plt.plot(dat[:,0], dat[:,1], label = "mean")
plt.plot(dat[:,0], dat[:,1]+dat[:,2], linestyle = "dashed", linewidth = 1, color = "black", label = "uncertainty")
plt.plot(dat[:,0], dat[:,1]-dat[:,2], linestyle = "dashed", linewidth = 1, color = "black")
plt.title("Random Walk in Discrete Lattice")
plt.legend()

plt.subplot(122)
#plt.plot(dat2[:,0], dat2[:,2]/dat2[:,1], label = "relative error") #for plotting the relative error
plt.plot(dat2[:,0], dat2[:,1], label = "mean")
plt.plot(dat2[:,0], dat2[:,1]+dat2[:,2], linestyle = "dashed", linewidth = 1, color = "black", label = "uncertainty")
plt.plot(dat2[:,0], dat2[:,1]-dat2[:,2], linestyle = "dashed", linewidth = 1, color = "black")
plt.title("Random Walk in Continous Lattice")
plt.legend()
plt.show()
```
