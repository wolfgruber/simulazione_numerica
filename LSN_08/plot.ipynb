```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

# Exercise 08.1:
minimizing the energy of a particle withthe wave function $
\Psi_T^{\sigma,\mu}(x) = e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$ with variational Monte Carlo:

```python
mi = np.loadtxt("min.dat")


plt.figure(figsize = (15,10))

plt.subplot(221)
plt.errorbar(np.arange(mi[:,0].size), mi[:,0], yerr=mi[:,1])
plt.title("Energy")

plt.subplot(222)
plt.plot(mi[:,2], mi[:,3])
plt.xlabel("$\mu$")
plt.ylabel("$\sigma$")
plt.title("$(\mu, \sigma)$")

plt.subplot(223)
plt.plot(mi[:,2])
plt.title("$\mu$")

plt.subplot(224)
plt.plot(mi[:,3])
plt.title("$\sigma$")

plt.show()

print('The minimum energy is', mi[-1,0],'\pm', mi[-1,1], 'at mu =', mi[-1,2], 'and sigma =', mi[-1,3])
```

```python
#plotting the function:

pl = np.loadtxt("plt.dat")
print(np.shape(pl))

x = np.linspace(0.35,1,np.size(pl[0,:])) # range sigma
y = np.linspace(0.5,1,np.size(pl[:,0])) #range mu
xp, yp = np.meshgrid(x,y)

fig = plt.figure( figsize=(18,6) )
ax = fig.add_subplot(122, projection='3d')

ax.plot_surface(xp,yp,pl,cmap='viridis')

plt.subplot(121)
plt.contour(x,y,pl,50)
plt.xlabel('$\sigma$')
plt.ylabel('$\mu$')

plt.show()
```

```python
#checking the last wave function

psi = np.loadtxt("psi.dat")

plt.figure(figsize=(12,4))

plt.subplot(121)
h = plt.hist(psi, 100)

plt.subplot(122)
plt.plot(h[1][0:-1], h[0]/np.sum(h[0]))
plt.show()
```
