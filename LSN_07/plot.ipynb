# Exercise 07.1:

```python
def var(a,b):
    c = np.mean(a*b) - np.mean(a)*np.mean(b)
    return c

def corr(a,b):
    m = np.size(a)
    c = np.zeros(m//2)
    
    for i in range(m//2):
        c[i] = var(a, np.roll(b,i))
    c = c/var(a,b)
    return c

def error(a):
    n = 100
    l = np.empty(2*n, dtype = int)
    er = np.zeros(np.size(l))
    p = 10**(1/n)

    for i in range(np.size(l)):
        l[i] = int(np.round_(10 * p**i))
        m = np.mean(a[0:l[i]])
        s = 0
        for j in range(l[i]):
            s += (m - a[j])**2
        er[i] = np.sqrt(s/(l[i]-1))
        
    return l, er
```

For the solid phase:

```python
ene = np.loadtxt("outtes.dat")
pre = np.loadtxt("outtps.dat")
n = ene[-1,0]

cene = corr(ene[:,1], ene[:,1])
cpre = corr(pre[:,1], pre[:,1])
x = np.array(range(int(n//2)))

te = np.polyfit(range(3),np.log(cene[0:3]),1)
tp = np.polyfit(range(3),np.log(cpre[0:3]),1)

exe = np.exp(x*te[0])
expr = np.exp(x*tp[0])

print(-1/te[0]+te[1])
print(-1/tp[0]+tp[1])


plt.figure(figsize = (15,10))

plt.subplot(221)
plt.plot(ene[:,0], ene[:,1])

plt.subplot(222)
plt.plot(pre[:,0], pre[:,1])

plt.subplot(223)
plt.plot(cene)
plt.plot(exe)
#plt.xlim(0,50)

plt.subplot(224)
plt.plot(cpre)
plt.plot(expr)

plt.show()
```
