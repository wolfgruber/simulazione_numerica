```python
import numpy as np
import matplotlib.pyplot as plt
```

# Exercise 10.1:
simulated annealing:

```python
cm = np.loadtxt("cmin.dat")
cconf = np.loadtxt("cconf.dat")
sm = np.loadtxt("smin.dat")
sconf = np.loadtxt("sconf.dat")

plt.figure(figsize = (15,10))

plt.subplot(221)
plt.plot(cm)
plt.title('Length of the Shortest path on the circle')

plt.subplot(222)
plt.plot(cconf[:,0], cconf[:,1])
plt.plot(cconf[:,0], cconf[:,1], 'o')
plt.title('Shortest path on the circle')

plt.subplot(223)
plt.plot(sm)
plt.title('Length of the Shortest path inside the square')

plt.subplot(224)
plt.plot(sconf[:,0], sconf[:,1])
plt.plot(sconf[:,0], sconf[:,1], 'o')
plt.title('Shortest path inside the square')

plt.show()
```

# Exercise 10.2:
simulated annealing with paralell computing:

```python
pm = np.loadtxt("paramin.dat")
pconf = np.loadtxt("paraconf.dat")

plt.figure(figsize = (15,5))

plt.subplot(121)
plt.plot(pm)
plt.title('Length of the Shortest path inside the square')

plt.subplot(122)
plt.plot(pconf[:,0], pconf[:,1])
plt.plot(pconf[:,0], pconf[:,1], 'o')
plt.title('Shortest path inside the square')

plt.show()
```
