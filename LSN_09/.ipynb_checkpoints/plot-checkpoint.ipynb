```python
import numpy as np
import matplotlib.pyplot as plt
```

# Exercise 09: 'Travelling salesman'
towns on a circle:

```python
m = np.loadtxt("cmin.dat")
mean = np.loadtxt("cmean.dat")
co = np.loadtxt("cconf.dat")

plt.figure(figsize = (15,5))

plt.subplot(121)
plt.plot(m, label='shortest route')
plt.plot(mean, label='mean over shortest half of routes')
plt.legend()
plt.title('Length of routes')

plt.subplot(122)
plt.plot(co[:,0], co[:,1])
plt.plot(co[:,0], co[:,1],'o')
plt.title('Shortest route through the cities')

plt.show()
```
