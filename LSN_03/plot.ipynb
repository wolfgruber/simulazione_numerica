```python
import numpy as np
from matplotlib import pyplot as plt
```

# Ex 03.1.1

Sampling the asset prize in one timestep.
$$S(T) = S(0) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)T
    +\sigma Z_{i+1}\sqrt{T}\right]$$

```python
c1 = np.loadtxt("call1.dat")
p1 = np.loadtxt("put1.dat")


print("Call:", c1[-1,0], "+/-", c1[-1,1])
print("Put: ", p1[-1,0], "+/-", p1[-1,1])

n = np.size(c1[:,0])
bsc = 14.976
bsp = 5.4595

plt.figure(figsize = (15,4))

plt.subplot(121)
plt.hlines(y = bsc, xmin = 0, xmax = n, linestyle = "dashed", color = "black", linewidth = 1, label = "b-s value")
plt.plot(c1[:,0], label = "call")
plt.plot(c1[:,0] + c1[:,1], linestyle = "dashed", color = "red", linewidth = 0.5, label = "$\pm \sigma$")
plt.plot(c1[:,0] - c1[:,1], linestyle = "dashed", color = "red", linewidth = 0.5)
plt.ylim(12,22)
plt.legend()
plt.title("Call")
plt.xlabel("n")


plt.subplot(122)
plt.hlines(y = bsp, xmin = 0, xmax = n, linestyle = "dashed", color = "black", linewidth = 1, label = "b-s value")
plt.plot(p1[:,0], label = "put")
plt.plot(p1[:,0] + p1[:,1], linestyle = "dashed", color = "red", linewidth = 0.5, label = "$\pm \sigma$")
plt.plot(p1[:,0] - p1[:,1], linestyle = "dashed", color = "red", linewidth = 0.5)
plt.ylim(1.5,8.5)
plt.legend()
plt.title("Put")
plt.xlabel("n")

plt.show()
```

# Ex 03.1.2

Sampling the asset prizes with $m = 100$ discrete timesteps.
$$S(T) = S(0) \prod_{i = 1}^{m} \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)t
    +\sigma Z_{i+1}\sqrt{t}\right] ~,~ t = \frac{T}{m}$$

```python
c2 = np.loadtxt("call2.dat")
p2 = np.loadtxt("put2.dat")

print("Call:", c2[-1,0], "+/-", c2[-1,1])
print("Put: ", p2[-1,0], "+/-", p2[-1,1])

plt.figure(figsize = (15,4))

plt.subplot(121)
plt.plot(c2[:,0], label = "call")
plt.hlines(y = bsc, xmin = 0, xmax = n, linestyle = "dashed", color = "black", linewidth = 1, label = "b-s value")
plt.plot(c2[:,0] + c2[:,1], linestyle = "dashed", color = "red", linewidth = 0.5, label = "$\pm \sigma$")
plt.plot(c2[:,0] - c2[:,1], linestyle = "dashed", color = "red", linewidth = 0.5)
plt.ylim(10,19)
plt.legend()
plt.xlabel("n")

plt.subplot(122)
plt.plot(p2[:,0], label = "put")
plt.hlines(y = bsp, xmin = 0, xmax = n, linestyle = "dashed", color = "black", linewidth = 1, label = "b-s value")
plt.plot(p2[:,0] + p2[:,1], linestyle = "dashed", color = "red", linewidth = 0.5, label = "$\pm \sigma$")
plt.plot(p2[:,0] - p2[:,1], linestyle = "dashed", color = "red", linewidth = 0.5)
plt.ylim(2,9)
plt.legend()
plt.xlabel("n")

plt.show()
```
